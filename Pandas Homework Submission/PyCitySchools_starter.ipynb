```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
school_data_to_load = "./schools_complete.csv"
student_data_to_load = "./students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#the following are a breakdown of each component of a final summary table
#this counts the total number of schools by measuring the length of the column school_names
school_count = len(school_data["school_name"])
school_count
```

```python
#this counts the total number of students by measuring adding the school sizes together
total_students = school_data["size"].sum()
total_students
```

```python
#this counts the total district budget by adding the individual school budgets together
total_budget = school_data["budget"].sum()
total_budget
```

```python
#this calculates the average math score for the district by calculating the total math scores divided by the total number of students
avg_math = school_data_complete["math_score"].sum() / total_students
avg_math.round(2)
```

```python
#same as above but for reading
avg_read = school_data_complete["reading_score"].sum() / total_students
avg_read.round(2)
```

```python
#Calculates the total pass average by finding avg of math and reading
pass_rate = (avg_math + avg_read) / 2
pass_rate.round(2)
```

```python
#calculates the total number of students with ,ath score above or equal to 70
pass_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
total_pass_math = pass_math["math_score"].count()
#calculates the percent by adding total # of passing students divided by total students
percent_pass_math  = (total_pass_math / total_students) *100
percent_pass_math ="{:.2f} %".format(percent_pass_math)
percent_pass_math
```

```python
#same as above but for reading
pass_read = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
total_pass_read = pass_read["reading_score"].count()
percent_pass_read  = (total_pass_read / total_students) *100
percent_pass_read ="{:.2f} %".format(percent_pass_read)
percent_pass_read
```

```python
#Creates summary table of the district data calculated in the previous 8 cells
index = ["Total Schools","Total Students","Total Budget","AVG Math Score","AVG Reading Score",
         "Pass Rate","% Students Pass Math","% Students Pass Reading"]
district_stats = pd.Series([school_count,total_students,total_budget,avg_math.round(2),avg_read.round(2),
               pass_rate.round(2),percent_pass_math,percent_pass_read],index= index)
district_summary = pd.DataFrame(district_stats,columns = ["District Stats"])
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#school summary requires the use of the school data database and this gives me an idea of its arrangement
school_data.head()
```

```python
#calculate a per student budget by dividing the total budget per school b y size of school 
school_data["Budget Per Student"] = school_data["budget"] / school_data["size"]
school_data.head()
```

```python
#This calculates the mean of student math scores per school
school_math_avg = school_data_complete.groupby("school_name")["math_score"].mean()

#merge math score average to the dataframe 
school_datav1 = school_data.merge(school_math_avg.to_frame(), left_on='school_name', right_index=True)
school_datav1.head()
```

```python
#same as above but for reading scores
school_read_avg = school_data_complete.groupby("school_name")["reading_score"].mean()
school_datav2 = school_datav1.merge(school_read_avg.to_frame(), left_on='school_name', right_index=True)

#rename columns
school_datav2 = school_datav2.rename(index=str, columns={"math_score": "AVG Math Score", "reading_score": "AVG Reading Score"})
school_datav2.head()
```

```python
#calculates the number of students that have a passing math score
school_pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['math_score'].count()

#merge that to data frame
school_datav3 = school_datav2.merge(school_pass_math.to_frame(), left_on='school_name', right_index=True)

#calculate the % passing math score total pass/total students
school_datav3["% Pass Math"] = (school_datav3["math_score"] / school_datav3["size"]) *100


```

```python
#same as above for reading
school_pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['reading_score'].count()
school_datav4 = school_datav3.merge(school_pass_read.to_frame(), left_on='school_name', right_index=True)
school_datav4["% Pass Read"] = (school_datav4["reading_score"] / school_datav4["size"]) *100

#rename the columns
school_datav4 = school_datav4.rename(index=str, columns={"math_score": "Total Pass Math", "reading_score": "Total Pass Read"})

#calculate overall passing by average of math and reading
school_datav4["% Overall Passing"] = (school_datav4["% Pass Math"] + school_datav4["% Pass Read"])/2
school_datav4.head()
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#this builds an array of info based on grade they are in
fresh =  school_data_complete.loc[school_data_complete["grade"] == "9th"]
soph =  school_data_complete.loc[school_data_complete["grade"] == "10th"]
junior =  school_data_complete.loc[school_data_complete["grade"] == "11th"]
senior =  school_data_complete.loc[school_data_complete["grade"] == "12th"]

#finds the average of math scores of each school based on school name
FRESH = fresh.groupby("school_name")["math_score"].mean()
SOPH = soph.groupby("school_name")["math_score"].mean()
JUNIOR = junior.groupby("school_name")["math_score"].mean()
SENIOR = senior.groupby("school_name")["math_score"].mean()

#merges all of the data together into 1 dataframe (Note: had to split into parts cause merge only allows to merge two things at a time)
math_by_grade = pd.merge(FRESH,SOPH, on = "school_name")
math_by_grade1 = pd.merge(JUNIOR,SENIOR, on="school_name")
math_by_grade2 = pd.merge(math_by_grade, math_by_grade1, on="school_name")

#rename all of the column headers
math_by_grade2.rename(index=str, columns={"math_score_x_x": "9th", "math_score_y_x": "10th",
                                          "math_score_x_y": "11th", "math_score_y_y": "12th"})
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
#same as above.  dataframe array for each grade already exists so no need to redo

#finds the average of reading scores of each school based on school name
rFRESH = fresh.groupby("school_name")["reading_score"].mean()
rSOPH = soph.groupby("school_name")["reading_score"].mean()
rJUNIOR = junior.groupby("school_name")["reading_score"].mean()
rSENIOR = senior.groupby("school_name")["reading_score"].mean()

#merges all of the data together into 1 dataframe
read_by_grade = pd.merge(rFRESH,rSOPH, on = "school_name")
read_by_grade1 = pd.merge(rJUNIOR,rSENIOR, on="school_name")
read_by_grade2 = pd.merge(read_by_grade, read_by_grade1, on="school_name")

#rename all of the column headers
read_by_grade2.rename(index=str, columns={"reading_score_x_x": "9th", "reading_score_y_x": "10th",
                                          "reading_score_x_y": "11th", "reading_score_y_y": "12th"})
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Sample bins. Feel free to create your own bins.
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
```

```python
#collects a dataframe with only my columns of interest
group_by_spending = school_datav4[["school_name","Budget Per Student","AVG Math Score", "AVG Reading Score",
                                 "% Pass Math","% Pass Read","% Overall Passing"]]

#bins the data based on budget per student and makes new column titled spending range per student
group_by_spending["Spending Ranges(per student)"] = pd.cut(group_by_spending["Budget Per Student"], spending_bins, labels=group_names)

#groups the data based on the spending range per student and finds the mean of each category of interest
group_by_spending=group_by_spending.groupby("Spending Ranges(per student)").agg({"AVG Math Score":"mean",
                                                                "AVG Reading Score":"mean","% Pass Math":"mean","% Pass Read":"mean",
                                                                "% Overall Passing":"mean"})

group_by_spending
```
